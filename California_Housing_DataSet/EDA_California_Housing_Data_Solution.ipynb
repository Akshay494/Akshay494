## 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

## **Solution 1:**

```python
dataset = pd.read_excel("/content/Dataset.xlsx")
dataset.head(5)
```

```python
# dataset["total_bedrooms_Size"] = 0
```

```python
dataset.head(5)
```

```python
## Creating new column on the basis of segregating the data on Small, Medium and Large category on total bedrooms into a list using a for loop.
list_bedrooms_Size=[]
for n in dataset["total_bedrooms"]:
  if n < 10.0:
    # print("small")
    list_bedrooms_Size.append("Small")
  elif 10.0 < n < 1000.0:
    # print ("medium")
    list_bedrooms_Size.append("Median")
  else:
    # print ("large")
    list_bedrooms_Size.append("Large")

print(list_bedrooms_Size)
```

```python
dataset["total_bedroom_size"] = list_bedrooms_Size  ### Creating new column based on the data from list.
```

```python
dataset.head(5)
```

```python
dataset.iloc[73]
```

```python
dataset.loc[70:75,"total_bedrooms":"total_bedroom_size":6]
```

```python
dataset.drop("total_bedroom_size", axis=1, inplace=True)
```
